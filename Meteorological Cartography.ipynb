```python
import numpy as np 
from matplotlib import pyplot as plt  # plotting function we will use today
import matplotlib.pyplot as plt       # achieves the same as the line above
import pandas as pd
```

```python
pathin      = '/Volumes/eas-shared/eas2900/2021/data_2021/' # path to data file
filein      = 'prelim_data.txt' # file name
filename    = pathin+filein # concatenate path and file name into one string
```

```python
dt = np.loadtxt(filename, delimiter=',') # load the text
```

```python
Day = range(1,32) # day for a month
T_high = dt[:,0] # the first column high temperature
T_low = dt[:,1] # the second column low temperature
T_ave = (T_high+T_low)/2 # the average temperature
GDD = T_ave-54 # GDD formula
Precip = dt[:,2] # the third column precipitation
```

```python
GDDtot = 0 # use to calculate the sum of GDD
Preptot = 0.00 # use to calculate the sum of Prep
```

```python
print('%-2s' % 'Day', '%-5s' % 'Tmax(F)', '%-5s' % 'Tmin(F)','%-5s' % 'Tave(F)','%-5s' % 'GDD(F)','%-5s' % 'Precip(inches)')    
for i in range(31):
    if GDD[i]<0:
        GDD[i] = 0 # if GDD<0 change it to 0
    GDDtot = GDDtot + GDD[i]
    Preptot = Preptot + Precip[i] # total calculation

for i in range(0,31):
    print('%2i' %Day[i],'%7.1f'% T_high[i],'%6.1f'% T_low[i],'%7.1f'% T_ave[i],'%6.1f'% GDD[i],'%11.2f'% Precip[i])
# print all the information needed
print('\n')
print('Total number of GDD ='+str(GDDtot)+' F')
print('Total precipitation ='+str('%.2f'% Preptot)+' inches')
print('\n')
# judge the GDDtot through criteria
if GDDtot<100:
    print('Corn is in bad shape')
elif GDDtot<200:
    print('Corn is in good shape')
else:
    print('Corn is in very good shape')
```

```python
fig = plt.figure(figsize=(10,7),facecolor='lightgray')
ax = fig.add_subplot(111) # background

ax.plot(GDD,color='b',linewidth=2.5,label='GDD') # GDD Data
ax.legend()
ax.grid()
ax.set_xlabel('Time (Day)')
ax.set_ylabel('Growing Degree Days ($^\circ$F)')
ax.set_title('Growing Degree Days (GDD)',fontsize=14)
ax.set_title('EAS2900 Prelim',fontsize=10,loc='left')
ax.set_title('Yitong Pan | yp258',fontsize=10,loc='right')
```
